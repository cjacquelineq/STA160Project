### Plot of Voting Trends over years

```python
import matplotlib.pyplot as plt
```

```python
D_years = [D_vote_08,D_vote_10,D_vote_12,D_vote_14]
R_years = [R_vote_08,R_vote_10,R_vote_12,R_vote_14]
years = ['2008','2010','2012','2014']

plt.plot(years, D_years, marker='o',label='Democratic', color='b')
plt.plot(years, R_years, marker='o', linestyle='--', color='r', label='Republic')
plt.xlabel('Year')
plt.ylabel('Votes')
plt.title('North Carolina CD12 Votes of Parties over Years (2008-2014)')
plt.legend()
plt.show()
```
